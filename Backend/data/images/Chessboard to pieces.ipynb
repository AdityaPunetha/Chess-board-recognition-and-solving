# Import Dependencies

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Get Image

```python
img = cv2.imread('chessboard\chessboard_all_pieces.png')
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
img.shape
```

```python
864/8
```

as you can see , the image is of shape 864x864 so when we divide it by 8 we will get cells of dimension 108x108

# Extracting cells

```python
def extract_cells(img):
    rows = np.vsplit(img, 8)
    boxes = []
    for r in rows:
        cols = np.hsplit(r, 8)
        for box in cols:
            boxes.append(box)
    return boxes
```

```python
cells = extract_cells(img)
```

```python
plt.imshow(cv2.cvtColor(cells[1], cv2.COLOR_BGR2RGB))
```

here you can also resize your image or transform it into gray scale if you would like but we will use app.roboflow.com for this purpose

# Saving images

```python
count = 0
for cell in cells:
    cv2.imwrite('Pieces/' + str(count) + '.png', cell)
    count += 1
```

This will crop and write all the individual cells into the pieces folder, repeat the process for other images but make sure to change the count value so it may not be overridden (you can also use uuid)
